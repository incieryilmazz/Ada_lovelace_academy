```python
base_url = "C:/Users/incie/Desktop/Ada_lovelace_academy/exercises/spotify/"
spotify_dataset_url = base_url + "songs_normalize.csv"
spotify_df = pd.read_csv(spotify_dataset_url)
```

```python
sns.set(style="darkgrid")
sns.histplot(data=spotify_df,
            x="tempo",
            kde=True,
            color="red"
        )
```

```python
grouped_df = spotify_df.groupby(['year', 'genre']).count().reset_index()
```

```python
grouped_df.head()
```

```python
hip_hop=grouped_df[grouped_df["genre"]=="hip hop"]
```

```python
hip_hop.head()
```

```python
plt.boxplot(x="year",data=hip_hop)  # bu grafikten çıkardığım sonuç ; sadece hip hop türü en çok 2007-2010 yılları arasında dinlenmiştir
```

```python
spotify_df.plot(kind='scatter', x='popularity', y='energy',color="purple")
plt.xlabel('popülerlik')
plt.ylabel('enerjileri')
plt.legend()
plt.grid(True)
plt.show()
```

```python
sns.regplot(
    x=spotify_df["energy"],
    y=spotify_df["danceability"],
    marker="+",
    line_kws={
        "color":"red",
        "alpha":0.6,
        "lw":5
    },
    scatter_kws={
        "color":"black",
        "s":50
    }

)
```

```python
sns.lmplot(
    x="speechiness", #konuşkanlık
    y="acousticness", #akustiklik
    hue="year",
    data=spotify_df
)
```

```python
sns.set(style="darkgrid")
sns.histplot(data=spotify_df,
            x="year",
            kde=True,
            color="darkcyan"
        )
```

```python
plt.barh(
    y=spotify_df.explicit,
    width=spotify_df.key,
    color=["orange","lightblue"],
    edgecolor="black"
)
```
