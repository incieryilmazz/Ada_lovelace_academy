```python
print(infected_df.columns.get_loc("Country/Region"))
print(recovered_df.columns.get_loc("Country/Region"))
print(deaths_df.columns.get_loc("Country/Region"))
```

```python
grouped_infected = infected_df.groupby('Country/Region').sum().iloc[:, 3:].sum(axis=1) # 3.sÃ¼tundan sonraki sÃ¼tunlarÄ± seÃ§er (3 dahil deÄŸil)
grouped_recovered = recovered_df.groupby('Country/Region').sum().iloc[:, 3:].sum(axis=1)
grouped_deaths = deaths_df.groupby('Country/Region').sum().iloc[:, 3:].sum(axis=1)
```

```python
result_df = pd.DataFrame({
    'Country/Region': grouped_infected.index,
    'Total Infected': grouped_infected.values,
    'Total Recovered': grouped_recovered.values,
    'Total Deaths': grouped_deaths.values
})
```
