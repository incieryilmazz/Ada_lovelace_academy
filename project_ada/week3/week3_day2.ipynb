```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.patches as mpatches
```

```python
df=pd.DataFrame({
    "Group" : ["A","B","C","D","E"],
    "Value" : [1,5,4,3,9]
})
```

```python
df=df.sort_values(by=["Value"])
plt.barh(
    y=df.Group,
    width=df.Value,
    color=["purple","pink","lightgreen","lightblue","lightyellow"],
    edgecolor="black"
    )
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.set(style="darkgrid")
sns.barplot(
    x="total_bill",
    y="day",
    data=tips,
    color="darkcyan"
)
plt.figure(figsize=(12,6))
```

```python
sns.barplot(
    x="tip",
    y="sex",
    data=tips,
    palette=["darkblue","red"]
)
plt.figure(figsize=(10,5))
```

```python
total=tips.groupby("day")["total_bill"].sum().reset_index()
```

```python
total
```

```python
smoker=tips[tips.smoker=='Yes']
```

```python
bar2=sns.barplot(
    x="day",
    y="total_bill",
    data=smoker,
    estimator=sum,
    errorbar=None,
    color="blue" 
)
```

```python
bar1=sns.barplot(x="day", y="total_bill", data=total, color="darkblue")
```

```python
bar1 = sns.barplot(x="day",  y="total_bill", data=total, color='darkblue')
bar2 = sns.barplot(
    x='day',
    y='total_bill',
    data=smoker,
    estimator=sum,
    errorbar=None,
    color='lightblue'
)
top_bar = mpatches.Patch(color='darkblue', label='non-smoker')
bottom_bar = mpatches.Patch(color='lightblue', label='smoker = Yes')
plt.legend(handles=[top_bar, bottom_bar])
plt.show()
```

```python
df_iris=sns.load_dataset("iris")
```

```python
df_iris.head()
```

```python
sns.histplot(data=df_iris,x="sepal_length",kde=True)
```

```python
sns.regplot(x=df_iris["sepal_length"], y=df_iris["sepal_width"])
```

```python
sns.regplot(
    x=df_iris["sepal_length"], 
    y=df_iris["sepal_width"],
    marker="+",
    line_kws={
        "color":"red",
        "alpha":0.6,
        "lw":5
    },
    scatter_kws={
        "color":"darkblue",
        "s":50
    }
    )
```

```python
sns.lmplot(
    x="sepal_length", 
    y="sepal_width",
    data=df_iris,
    hue="species",
    # fit_reg=False lineları kaldırıyor
)
```

```python
sns.boxplot(
    x=df_iris["species"],
    y=df_iris["sepal_length"]
)
```
